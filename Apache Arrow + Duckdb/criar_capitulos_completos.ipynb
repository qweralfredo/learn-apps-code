## üì¶ Imports e Configura√ß√£o

```python
import os
import json
from pathlib import Path
```

## üìã Defini√ß√£o da Estrutura dos Cap√≠tulos

```python
# Definir estrutura dos cap√≠tulos
capitulos_info = [
    ("02", "integracao_duckdb_arrow", "Integra√ß√£o DuckDB + Arrow"),
    ("03", "arrow_tables_datasets", "Arrow Tables e Datasets"),
    ("04", "zero_copy_performance", "Zero-Copy e Performance"),
    ("05", "arrow_flight_sql", "Arrow Flight SQL"),
    ("06", "streaming_batches", "Streaming e Batches"),
    ("07", "ipc_serializacao", "Arrow IPC e Serializa√ß√£o"),
    ("08", "integracao_pandas_polars", "Integra√ß√£o Pandas e Polars"),
    ("09", "compute_functions", "Arrow Compute Functions"),
    ("10", "casos_uso_otimizacoes", "Casos de Uso e Otimiza√ß√µes")
]

output_dir = Path(r"C:\projetos\Cursos\Apache Arrow + Duckdb\code")

print(f"üìÅ Diret√≥rio de sa√≠da: {output_dir}")
print(f"üìö Total de cap√≠tulos: {len(capitulos_info)}")
```

## üêç Template para Arquivos Python

```python
def create_python_file(num, nome, titulo, output_dir):
    """Cria arquivo Python com conte√∫do de exemplo."""
    py_content = f'''# -*- coding: utf-8 -*-
"""
Cap√≠tulo {num}: {titulo}
Curso: Apache Arrow + DuckDB
"""

import sys
import io
sys.stdout = io.TextIOWrapper(sys.stdout.buffer, encoding='utf-8')

import pyarrow as pa
import duckdb
import pandas as pd
import numpy as np

print("="*60)
print("CAP√çTULO {num}: {titulo.upper()}")
print("="*60)

# Criar dados de exemplo
n = 100_000
data = pa.table({{
    'id': range(n),
    'value': [i * 2.5 for i in range(n)],
    'category': [f'Cat_{{i%10}}' for i in range(n)]
}})

print(f"\\nDados: {{n:,}} linhas")

# Conectar DuckDB
con = duckdb.connect()

# Query de exemplo
result = con.execute("""
    SELECT
        category,
        count(*) as count,
        avg(value) as avg_value,
        sum(value) as sum_value
    FROM data
    GROUP BY category
    ORDER BY category
""").arrow()

print("\\nResultado:")
print(result.to_pandas())

print("\\n" + "="*60)
print("CAP√çTULO {num} CONCLU√çDO")
print("="*60)
'''
    
    py_file = output_dir / f"capitulo_{num}_{nome}.py"
    with open(py_file, 'w', encoding='utf-8') as f:
        f.write(py_content)
    
    return py_file

print("‚úÖ Fun√ß√£o create_python_file() definida")
```

## üìì Template para Notebooks Jupyter

```python
def create_notebook_file(num, nome, titulo, output_dir):
    """Cria notebook Jupyter com estrutura completa."""
    nb_cells = [
        {
            "cell_type": "markdown",
            "metadata": {},
            "source": [
                f"# Cap√≠tulo {num}: {titulo}\n",
                "## Curso: Apache Arrow + DuckDB"
            ]
        },
        {
            "cell_type": "code",
            "execution_count": None,
            "metadata": {},
            "outputs": [],
            "source": [
                "# !pip install pyarrow duckdb pandas"
            ]
        },
        {
            "cell_type": "code",
            "execution_count": None,
            "metadata": {},
            "outputs": [],
            "source": [
                "import pyarrow as pa\n",
                "import duckdb\n",
                "import pandas as pd\n",
                "import numpy as np\n",
                "\n",
                f"print('Cap√≠tulo {num}: {titulo}')"
            ]
        },
        {
            "cell_type": "markdown",
            "metadata": {},
            "source": ["## Dados de Exemplo"]
        },
        {
            "cell_type": "code",
            "execution_count": None,
            "metadata": {},
            "outputs": [],
            "source": [
                "n = 100_000\n",
                "data = pa.table({\n",
                "    'id': range(n),\n",
                "    'value': [i * 2.5 for i in range(n)],\n",
                "    'category': [f'Cat_{i%10}' for i in range(n)]\n",
                "})\n",
                "\n",
                "print(f'Dados: {n:,} linhas')\n",
                "print(data.schema)"
            ]
        },
        {
            "cell_type": "markdown",
            "metadata": {},
            "source": ["## Query com DuckDB"]
        },
        {
            "cell_type": "code",
            "execution_count": None,
            "metadata": {},
            "outputs": [],
            "source": [
                "con = duckdb.connect()\n",
                "\n",
                "result = con.execute('''\n",
                "    SELECT\n",
                "        category,\n",
                "        count(*) as count,\n",
                "        avg(value) as avg_value\n",
                "    FROM data\n",
                "    GROUP BY category\n",
                "    ORDER BY category\n",
                "''').arrow()\n",
                "\n",
                "result.to_pandas()"
            ]
        }
    ]

    notebook = {
        "cells": nb_cells,
        "metadata": {
            "kernelspec": {
                "display_name": "Python 3",
                "language": "python",
                "name": "python3"
            },
            "language_info": {
                "name": "python",
                "version": "3.8.0"
            }
        },
        "nbformat": 4,
        "nbformat_minor": 4
    }
    
    nb_file = output_dir / f"capitulo_{num}_{nome}.ipynb"
    with open(nb_file, 'w', encoding='utf-8') as f:
        json.dump(notebook, f, indent=1)
    
    return nb_file

print("‚úÖ Fun√ß√£o create_notebook_file() definida")
```

## üöÄ Execu√ß√£o: Cria√ß√£o dos Arquivos

```python
# Criar todos os cap√≠tulos
for num, nome, titulo in capitulos_info:
    # Criar arquivo Python
    py_file = create_python_file(num, nome, titulo, output_dir)
    print(f"‚úÖ Python: {py_file.name}")
    
    # Criar notebook
    nb_file = create_notebook_file(num, nome, titulo, output_dir)
    print(f"‚úÖ Notebook: {nb_file.name}")
    print()
```
