## üêç Template Avan√ßado Python

```python
python_template = '''# -*- coding: utf-8 -*-
"""
Cap√≠tulo {num}: {titulo}
Curso: Apache Arrow + DuckDB

Configura√ß√£o para Windows (UTF-8)
"""

import sys
import io
sys.stdout = io.TextIOWrapper(sys.stdout.buffer, encoding='utf-8')

# Instalar depend√™ncias
print("# pip install pyarrow duckdb pandas polars")

import pyarrow as pa
import duckdb
import pandas as pd
import numpy as np

print("\\n" + "="*60)
print("CAP√çTULO {num}: {titulo_upper}")
print("="*60)

# Conectar DuckDB
con = duckdb.connect()

# Criar dados de exemplo
n = 100_000
sample_data = pa.table({{
    'id': range(n),
    'value': [i * 2.5 for i in range(n)],
    'category': [f'Cat_{{i%10}}' for i in range(n)],
    'timestamp': pa.array(['2024-01-01'] * n)
}})

print(f"\\nDados de exemplo: {{n:,}} linhas")
print(sample_data.schema)

# SE√á√ÉO 1: {topico1}
print("\\n{num}.1 {topico1_upper}")
print("-"*60)

# Query b√°sica
result = con.execute("""
    SELECT
        category,
        count(*) as count,
        avg(value) as avg_value
    FROM sample_data
    GROUP BY category
    ORDER BY category
""").arrow()

print("\\nResultado da query:")
print(result.to_pandas())

# SE√á√ÉO 2: {topico2}
print("\\n{num}.2 {topico2_upper}")
print("-"*60)

# Convers√£o para Pandas
df_result = result.to_pandas()
print("\\nConvertido para Pandas:")
print(df_result.head())

# SE√á√ÉO 3: {topico3}
print("\\n{num}.3 {topico3_upper}")
print("-"*60)

# Integra√ß√£o com Polars (se dispon√≠vel)
try:
    import polars as pl
    df_polars = pl.from_arrow(sample_data)
    print("\\nConvertido para Polars:")
    print(df_polars.head())
except ImportError:
    print("\\nPolars n√£o instalado (opcional)")

print("\\n" + "="*60)
print("CAP√çTULO {num} CONCLU√çDO!")
print("="*60)
print("\\nVoc√™ aprendeu:")
{aprendizados}
'''

print("‚úÖ Template Python avan√ßado definido")
```
